```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.barplot()
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill,)
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill,hue=tips_df.sex,)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df)
```

```python
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,order=order)
```

```python
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,hue_order=hue_order)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,hue_order=hue_order,estimator=np.mean)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,hue_order=hue_order,estimator=np.max)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,hue_order=hue_order,ci=55)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,estimator=np.max,n_boot=2)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,orient='v')
```

```python
sns.barplot(y='day',x='total_bill',hue='sex',
           data=tips_df)
```

```python
sns.barplot(y='size',x='total_bill',hue='sex',
           data=tips_df,orient='h')
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,color='g',palette='magma',saturation=0.3)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,errcolor='0.1')
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,errwidth=12)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,capsize=1)
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,dodge=False)
```

```python
kwargs={'alpha':0.9,'linestyle':':','linewidth':5,"edgecolor":'k'}
sns.barplot(x='day',y='total_bill',hue='sex',
           data=tips_df,**kwargs)
```

```python
sns.set()
kwargs={'alpha':0.9,'linestyle':':','linewidth':5,"edgecolor":'k'}
sns.barplot(x='day',y='total_bill',
           data=tips_df,**kwargs)
```

```python
sns.barplot(x='day',y='total_bill',
           data=tips_df,alpha=0.9,linestyle="-",linewidth=3,edgecolor="g")
```

```python
ax=sns.barplot(x='day',y='total_bill',
           data=tips_df,alpha=0.9,linestyle="-",linewidth=3,edgecolor="g")
ax.set(title="Bar plot of tips DF",
      xlabel="Days",ylabel="Total bill")
```

```python
plt.figure(figsize=(16,6))
ax=sns.barplot(x='day',y='total_bill',
           data=tips_df,alpha=0.9,linestyle="-",linewidth=3,edgecolor="g")
ax.set(title="Bar plot of tips DF",
      xlabel="Days",ylabel="Total bill")
plt.title("Barplot of Days and Total Bill ",fontsize=30)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Total Bill",fontsize=20)
plt.savefig("seaborn_Barplot")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('D:\python programs\program\covid1.csv',nrows=5)
sns.barplot(x=df['total cases'],y=df['new cases'],hue=df['new deaths'],)
plt.show()



plt.figure(figsize=(9,4))
ax=sns.barplot(x='new cases',y='new deaths',
           data=df,alpha=0.9,linestyle="-",linewidth=3,edgecolor="g")
ax.set(title="Bar plot of tips DF",
      xlabel="new cases",ylabel="new deaths")
plt.title("Barplot of new cases and new deaths ",fontsize=30)
plt.xlabel("new cases",fontsize=20)
plt.ylabel("new deaths",fontsize=20)
plt.savefig("seaborn_Barplot")
plt.show()
```
