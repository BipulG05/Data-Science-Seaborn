```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.distplot(tips_df["size"])
```

```python
sns.distplot(tips_df["tip"])
```

```python
sns.distplot(tips_df["total_bill"],label="Total bill")
plt.title("Histogram of total bill")
plt.legend()
```

```python
plt.figure(figsize=(16,6))
sns.set()
sns.distplot(tips_df["total_bill"],label="Total Bill")
plt.title("Histrogram of Total Bill")
plt.legend()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,6))
sns.set()
sns.distplot(tips_df["total_bill"],bins=bins)
plt.xticks(bins)
plt.title("Histrogram of Total Bill")
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,6))
sns.set()
sns.distplot(tips_df["total_bill"],hist_kws={"color":'#DC143C',"edgecolor":'#aaff00',"linewidth":5,
            "linestyle":'--',"alpha":0.9}) # r="#DC143C"
plt.title("Histrogram of Total Bill")
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(16,6))
sns.set()
sns.distplot(tips_df["total_bill"],hist_kws={"color":'#DC143C',"edgecolor":'#aaff00',"linewidth":5,
            "linestyle":'--',"alpha":0.9},kde_kws={"color":'#8e00ce',"linewidth":12,
            "linestyle":'--',"alpha":0.9},rug=True,rug_kws={"color":'#0426d0',"linewidth":3,
            "linestyle":'--',"alpha":0.9},label="TB") # r="#DC143C"
plt.title("Histrogram of Total Bill",fontsize=22)
plt.xlabel("Tatal Bill",fontsize=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.distplot(tips_df["total_bill"],bins=9,label="total_bill")
sns.distplot(tips_df["tip"],bins=9,label="tip")
sns.distplot(tips_df["size"],bins=9,label="size")
plt.legend()
```
