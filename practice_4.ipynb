# Heatmap :part 2

```python
plt.figure(figsize=(16,6))

annot_kws={'fontsize':10,
          'fontstyle':'italic',
           'color':'k',
          'alpha':0.9,
          'rotation':'vertical',
          'verticalalignment':'center',
          'backgroundcolor':'w'}

sns.heatmap(Global_Warming_df,cmap='coolwarm',annot=True,annot_kws=annot_kws)
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(Global_Warming_df,cmap='turbo',annot=True,linewidth=4)
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(Global_Warming_df,cmap='turbo',annot=True,linewidth=4,linecolor='k')
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(Global_Warming_df,cmap='turbo',annot=True,linewidth=4,linecolor='k',cbar=False)
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(Global_Warming_df,cmap='turbo',annot=True,linewidth=4,linecolor='k',
            cbar=False,xticklabels=False,yticklabels=False)
```

```python
plt.figure(figsize=(16,6))

cbar_kws={'orientation':'horizontal',
         'shrink':1,
         'extend':'min',
         'extendfrac':0.1,
         'ticks':np.arange(0,22),
         'drawedges':True}

sns.heatmap(Global_Warming_df,cbar_kws=cbar_kws)
```

```python
plt.figure(figsize=(16,6))
country_code=['USA','GBR','IND','CHN','RUS','AUS','FRA','DEU','CAN','BRA','ARG','PAK']
sns.heatmap(Global_Warming_df,cmap='turbo',annot=True,linewidth=4,linecolor='k',
            cbar=False,xticklabels=np.arange(0,15),yticklabels=country_code)
```

```python
plt.figure(figsize=(16,6))
ax=sns.heatmap(Global_Warming_df)
ax.set(title="Heatmap of 'who is responsible for global warming '",
      xlabel="CO2 emissions (metric ton per captia) per year",
      ylabel="Country Name")
sns.set(font_scale=0.1)
```

# Heatmap :part 3

# Correlation in Heatmap

```python
Global_Warming_df.corr()
```

```python
plt.figure(figsize=(16,6))
ax=sns.heatmap(Global_Warming_df.corr(),annot=True,linewidth=3,)
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='k')
plt.title("Heatmap of 'who is responsible for global warming' ",fontsize=25)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
cancer_dataset
```

```python
cancer_df=pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
                      columns=np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize=(65,20))
ax=sns.heatmap(cancer_df.corr(),annot=True,linewidth=3)
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='k')
plt.title("Heatmap of Breast Cancer ",fontsize=45)
plt.show()
```
